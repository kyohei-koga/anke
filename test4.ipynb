```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.font_manager import FontProperties
%matplotlib inline

fp = FontProperties(fname=u'/usr/share/fonts/truetype/fonts-japanese-gothic.ttf',size=10)
```

```python
data_columns = pd.read_excel('/home/kyohei/NEDO_anke_data/NEDO_H24.xlsx').columns
data = pd.read_excel('/home/kyohei/NEDO_anke_data/NEDO_H24.xlsx',skiprows=6,skip_footer=12)
```

```python
fig ,ax = plt.subplots(3,2,figsize=(12,10))
axes = ax.flatten()

data[u'Q4_目標明確'].value_counts().sort_index().plot(kind='bar',ax=axes[0])
axes[0].set_xlim([-0.5,2.5])
axes[0].set_xticks([0,1,2])
axes[0].set_xticklabels([u'明確であった',u'ある程度は明確であった',u'不明確であった'],fontproperties=fp,rotation=45,fontsize=8)
axes[0].set_title(u'目的の明確さ',fontproperties=fp)
plt.subplots_adjust(hspace=0.7)

data[u'Q4_目標共有'].value_counts().sort_index().plot(kind='bar',ax=axes[1])
axes[1].set_xlim([-0.5,2.5])
axes[1].set_xticks([0,1,2])
axes[1].set_xticklabels([u'共有できていた',u'一部の機関とは共有できていた',u'まったく共有できていなかった'],fontproperties=fp,rotation=45,fontsize=8)
axes[1].set_title(u'目的の共有',fontproperties=fp)

data[u'Q4_目標明確'].value_counts().sort_index().plot(kind='bar',ax=axes[2])
axes[2].set_xlim([-0.5,2.5])
axes[2].set_xticks([0,1,2])
axes[2].set_xticklabels([u'明確であった',u'ある程度は明確であった',u'不明確であった'],fontproperties=fp,fontsize=8,rotation=45)
axes[2].set_title(u'目標の明確さ',fontproperties=fp)

data[u'Q4_目標共有'].value_counts().sort_index().plot(kind='bar',ax=axes[3])
axes[3].set_xlim([-0.5,2.5])
axes[3].set_xticks([0,1,2])
axes[3].set_xticklabels([u'共有できていた',u'一部の機関とは共有できていた',u'まったく共有できていなかった'],fontproperties=fp,fontsize=8,rotation=45)
axes[3].set_title(u'目標の共有',fontproperties=fp)

data[u'Q4_役割分担'].value_counts().sort_index().plot(kind='bar',ax=axes[4])
axes[4].set_xlim([-0.5,2.5])
axes[4].set_xticks([0,1,2])
axes[4].set_xticklabels([u'明確であった',u'ある程度は明確であった',u'不明確であった'],fontproperties=fp,fontsize=8,rotation=45)
axes[4].set_title(u'目標の共有',fontproperties=fp)

data[u'Q4_役割共有'].value_counts().sort_index().plot(kind='bar',ax=axes[5])
axes[5].set_xlim([-0.5,2.5])
axes[5].set_xticks([0,1,2])
axes[5].set_xticklabels([u'共有できていた',u'一部の機関とは共有できていた',u'まったく共有できていなかった'],fontproperties=fp,fontsize=8,rotation=45)
axes[5].set_title(u'役割分担の共有',fontproperties=fp)
```
